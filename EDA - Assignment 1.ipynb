```python
import pandas as pd

df = pd.read_csv(r"C:\Users\mmoammar\Downloads\COVID_clinical_trials.csv")

print(df)
```

```python
# Check the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)
print("------------------------------------")

# Display the columns in the DataFrame
print("\nColumns in the DataFrame:", df.columns)
print("------------------------------------")

# Display the first 5 rows
print("\nFirst 5 rows:")
print(df.head(5))
print("------------------------------------")

# Display the last 5 rows
print("\nLast 5 rows:")
print(df.tail(5))
print("------------------------------------")

# Display the first 15 rows
print("\nFirst 15 rows:")
print(df.head(15))
print("------------------------------------")

# Display the last 15 rows
print("\nLast 15 rows:")
print(df.tail(15))
print("------------------------------------")

# Display the first 30 rows
print("\nFirst 30 rows:")
print(df.head(30))
print("------------------------------------")

# Display the last 30 rows
print("\nLast 30 rows:")
print(df.tail(30))
print("------------------------------------")
```

```python
# Display information about the DataFrame, including data types and non-null values
print("DataFrame Information:")
df.info()
print("------------------------------------")

# Display descriptive statistics for numerical columns
print("\nDescriptive Statistics for Numerical Columns:")
print(df.describe())
print("------------------------------------")

# Display descriptive statistics for all columns, including non-numeric ones
print("\nDescriptive Statistics for All Columns:")
print(df.describe(include='all'))
print("------------------------------------")
```

```python
# Determine missing values in the whole dataset
total_missing = df.isnull().sum().sum()
print("Total Missing Values in the Whole Dataset:", total_missing)
print("------------------------------------")

# Analyze missing values in each column
missing_per_column = df.isnull().sum()
percent_missing_per_column = (missing_per_column / len(df)) * 100

# Create a DataFrame to display missing values per column
missing_info = pd.DataFrame({
    'Missing Values': missing_per_column,
    'Percentage Missing': percent_missing_per_column
})

print("\nMissing Values in Each Column:")
print(missing_info)
```

```python
# Identify and sum up duplicate rows
duplicate_sum = df.duplicated().sum()

print("Sum of Duplicate Rows:", duplicate_sum)
```

```python
# Question 1: What is the average enrollment for each study phase?
average_enrollment_by_phase = df.groupby('Phases')['Enrollment'].mean()

# Question 2: How many studies are there for each status?
study_count_by_status = df.groupby('Status')['Rank'].count()

# Question 3: What is the total enrollment for studies funded by each entity?
total_enrollment_by_fund = df.groupby('Funded Bys')['Enrollment'].sum()

# Question 4: How many studies have reported results by gender?
results_by_gender_count = df.groupby('Gender')['Rank'].count()

# Question 5: What is the maximum rank for each study type?
max_rank_by_study_type = df.groupby('Study Type')['Rank'].max()

# Displaying the results
print("1. Average Enrollment by Phase:")
print(average_enrollment_by_phase)

print("\n2. Study Count by Status:")
print(study_count_by_status)

print("\n3. Total Enrollment by Funded Entity:")
print(total_enrollment_by_fund)

print("\n4. Results Count by Gender:")
print(results_by_gender_count)

print("\n5. Maximum Rank by Study Type:")
print(max_rank_by_study_type)
```

```python
# Question 1: How many studies have enrollment greater than 1000?
studies_above_1000_enrollment = df[df['Enrollment'] > 1000].shape[0]

# Question 2: What is the average enrollment for studies with 'Phase 2'?
average_enrollment_phase_2 = df[df['Phases'] == 'Phase 2']['Enrollment'].mean()

# Question 3: How many studies have reported results and are in 'Recruiting' status?
recruiting_studies_with_results = df[(df['Study Results'] == 'Has Results') & (df['Status'] == 'Recruiting')].shape[0]

# Question 4: What is the minimum enrollment for studies funded by 'National Institutes of Health'?
min_enrollment_nih_funded = df[df['Funded Bys'] == 'National Institutes of Health']['Enrollment'].min()

# Question 5: How many studies have 'Interventional' study type and are in 'Completed' status?
completed_interventional_studies = df[(df['Study Type'] == 'Interventional') & (df['Status'] == 'Completed')].shape[0]

# Displaying the results
print("1. Studies with Enrollment > 1000:", studies_above_1000_enrollment)

print("\n2. Average Enrollment for Phase 2 Studies:", average_enrollment_phase_2)

print("\n3. Recruiting Studies with Results:", recruiting_studies_with_results)

print("\n4. Minimum Enrollment for NIH Funded Studies:", min_enrollment_nih_funded)

print("\n5. Completed Interventional Studies:", completed_interventional_studies)
```
